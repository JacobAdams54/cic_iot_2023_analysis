```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("../data/CIC_IOT_Dataset_2023/")
```

```python
# 1) Inventory files
files = sorted([p for p in DATA_DIR.rglob("*") if p.is_file()])
print("File count:", len(files))
for p in files[:50]:
    print(p.relative_to(DATA_DIR), p.stat().st_size)
```

```python
# 2) Find candidate tabular files
tabular = [p for p in files if p.suffix.lower() in [".csv", ".parquet", ".feather", ".tsv"]]
print("\nTabular files:", len(tabular))
for p in tabular[:50]:
    print(p.relative_to(DATA_DIR))
```

```python
# 3) Load ONE file sample safely (adjust once you see what exists)
if tabular:
    p = tabular[0]
    print("\nLoading sample from:", p.name)

    if p.suffix.lower() == ".csv":
        df = pd.read_csv(p, nrows=5000)
    elif p.suffix.lower() == ".tsv":
        df = pd.read_csv(p, sep="\t", nrows=5000)
    elif p.suffix.lower() in [".parquet", ".feather"]:
        df = pd.read_parquet(p)
        df = df.head(5000)
    else:
        df = None

    if df is not None:
        print("Shape (sample):", df.shape)
        display(df.head(3))
        display(df.dtypes)
        print("Missingness (top 20):")
        display(df.isna().mean().sort_values(ascending=False).head(20))
```
